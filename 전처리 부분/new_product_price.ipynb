```python
import pandas as pd
import glob
import os

def load_and_concat_csv_files(directory_path):
    # .csv 파일 찾기
    csv_files = glob.glob(os.path.join(directory_path, "*.csv"))

    # CSV 파일 확인
    if not csv_files:
        print("지정된 경로에 CSV 파일이 없습니다.")
        return None

    # CSV 파일을 읽어와 DataFrame으로 변환
    dataframes = []
    for file in csv_files:
        try:
            df = pd.read_csv(file, encoding='cp949')  # 인코딩 방식 cp949
            dataframes.append(df)
        except Exception as e:
            print(f"파일 {file} 읽기 중 오류 발생: {e}")
    
    # DataFrame 병합
    if dataframes:
        combined_df = pd.concat(dataframes, ignore_index=True)
        
        # 컬럼명 설정
        if len(combined_df.columns) >= 3:  # 최소 3개의 컬럼이 있어야 컬럼명을 변경 가능
            combined_df.columns = ["날짜", "예측 가격", "품목"]
        else:
            print("DataFrame에 충분한 컬럼이 없습니다. 컬럼명을 변경하지 않았습니다.")
        
        return combined_df
    else:
        print("CSV 파일을 병합할 수 없습니다.")
        return None

# 함수 호출 예제
directory_path = r"C:/Users/kitcoop/Desktop/LSTM결과/"
combined_df = load_and_concat_csv_files(directory_path)

# 결과 확인
if combined_df is not None:
    print("병합된 DataFrame 미리보기:")
    print(combined_df.head())
else:
    print("병합된 결과가 없습니다.")
```
