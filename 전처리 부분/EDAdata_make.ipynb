```python
df_oil = pd.read_csv('C:/Users/chany/OneDrive/바탕 화면/식품예측 데이터/주유소_평균판매가격_제품별_2014_2024.csv',encoding='cp949')
df_oil=df_oil.drop(columns='보일러등유')
df_oil=df_oil.rename(columns={'구분':'분류'})

def change_datetime(df, column_name):
    df[column_name] = pd.to_datetime(
        df[column_name]
        .str.replace("년", "-", regex=False)
        .str.replace("월", "-", regex=False)
        .str.replace("일", "", regex=False),
        format='%Y-%m-%d'
    )
    return df
df_oil = change_datetime(df_oil, "분류")
df_oil
```

```python
df= pd.merge(df, df_oil, how='left')
df
```

```python
df_wea = pd.read_csv('C:/Users/chany/OneDrive/바탕 화면/식품예측 데이터/날씨데이터_필터링.csv',encoding='cp949')
df_wea=df_wea.drop(columns='Unnamed: 0')
df_wea=df_wea.rename(columns={'관측일':'분류'})
df_wea['분류']=pd.to_datetime(df_wea['분류'], format='%Y-%m-%d')
```

```python
jijum_lst = {'밀양': 288, '대전': 133, '광주': 156, '인제': 211, '동두천': 98}
df_wea_dict = {} 

for k, v in jijum_lst.items():
    df_filter = df_wea[df_wea['지점번호'] == v]
    df_filter = df_filter.rename(columns={
        '일 평균기온': f'일 평균기온_{k}',
        '일교차': f'일교차_{k}',
        '일 강수량': f'일 강수량_{k}',
        '일 평균 상대습도': f'일 평균 상대습도_{k}',
    })
    df_wea_dict[k] = df_filter
```

```python
df_1=df_wea_dict['밀양'].drop(columns='지점번호')
df_2=df_wea_dict['대전'].drop(columns='지점번호')
df_3=df_wea_dict['광주'].drop(columns='지점번호')
df_4=df_wea_dict['인제'].drop(columns='지점번호')
df_5=df_wea_dict['동두천'].drop(columns='지점번호')
```
