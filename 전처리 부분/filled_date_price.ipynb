```python
import pandas as pd

file_path = '품목_소매가_전부_갱신.csv'
data = pd.read_csv(file_path,encoding='cp949')
data=data[data['품목']=='새송이버섯']
```

```python
data.columns = data.columns.str.strip()

data['구분'] = pd.to_datetime(data['구분'], errors='coerce')
data = data.sort_values('구분').reset_index(drop=True)

full_date_range = pd.date_range(start=data['구분'].min(), end=data['구분'].max())

data = data.set_index('구분').reindex(full_date_range).rename_axis('구분').reset_index()

data[['평균', '전통시장', '유통업체']] = data[['평균', '전통시장', '유통업체']].fillna(method='ffill')

display(data)
```

```python
data.to_csv("애호박_상품_1개_일별소매가.csv", index = False, encoding='cp949')
```

```python
# Rename the columns to remove any extra spaces
data.columns = data.columns.str.strip()

# Repeat the previous steps with the corrected column name
data['구분'] = pd.to_datetime(data['구분'], errors='coerce')
data = data.sort_values('구분').reset_index(drop=True)

# Create a full date range from the minimum to the maximum date in the data
full_date_range = pd.date_range(start=data['구분'].min(), end=data['구분'].max())

# Reindex the data to include the full date range, filling missing dates with NaN
data = data.set_index('구분').reindex(full_date_range).rename_axis('구분').reset_index()

# Fill missing prices with the most recent available price data
data[['평균', '전통시장', '유통업체']] = data[['평균', '전통시장', '유통업체']].fillna(method='ffill')

# Show the processed data
display(data)
```
