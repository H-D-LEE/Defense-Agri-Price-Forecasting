## 더미화 해서 랜덤포레스트로 작물 가격에 영향 큰 지점번호 찾으려 했으나 실패

```python
# 원-핫 인코딩
encoded_df = pd.get_dummies(merged_df, columns=['지점번호'])
encoded_df
```

```python
columns_with_지점번호 = encoded_df.columns[encoded_df.columns.str.contains('지점번호')]
columns_with_지점번호
```

```python
encoded_df[columns_with_지점번호] = encoded_df[columns_with_지점번호].astype(int)
encoded_df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# 피처와 타겟 설정
features = encoded_df.drop(columns=['전통시장'])
target = encoded_df['전통시장']

# 학습/테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 모델 학습
model = RandomForestRegressor()
model.fit(X_train, y_train)

# 피처 중요도 출력
importances = model.feature_importances_
print(f"지점번호 관련 피처 중요도: {importances}")
```
