```python
# 재배기간(+수확기간)과 수확기간 피처 엔지니어링(월/일 기준으로 설정, 연도는 고려하지 않음)

# 재배기간(+수확기간)
# 재배기간_1 봄배추
plant_start_month_day_1 = '04-01'
plant_end_month_day_1 = '07-20'

# 재배기간_2 가을배추
plant_start_month_day_2 = '08-21'
plant_end_month_day_2 = '12-20'

# 수확기간
# 수확기간_1 봄배추
harvest_start_month_day_1 = '07-11'
harvest_end_month_day_1 = '07-20'

# 수확기간_2 가을배추
harvest_start_month_day_2 = '11-01'
harvest_end_month_day_2 = '12-20'

# 연도를 임의로 지정 (예: 2000년)하여 월/일을 날짜로 변환
def convert_to_datetime(month_day_str):
    return pd.to_datetime(f"2000-{month_day_str}", format="%Y-%m-%d")

# 재배기간
# 월/일을 datetime 객체로 변환
plant_start_date_1 = convert_to_datetime(plant_start_month_day_1)
plant_end_date_1 = convert_to_datetime(plant_end_month_day_1)

plant_start_date_2 = convert_to_datetime(plant_start_month_day_2)
plant_end_date_2 = convert_to_datetime(plant_end_month_day_2)

# 재배기간 컬럼 생성 (월일 기준으로 비교)
df['재배기간_1'] = (df['날짜'].dt.strftime('%m-%d') >= plant_start_date_1.strftime('%m-%d')) & (df['날짜'].dt.strftime('%m-%d') <= plant_end_date_1.strftime('%m-%d'))
df['재배기간_2'] = (df['날짜'].dt.strftime('%m-%d') >= plant_start_date_2.strftime('%m-%d')) & (df['날짜'].dt.strftime('%m-%d') <= plant_end_date_2.strftime('%m-%d'))

# 재배기간 컬럼을 True/False에서 1/0으로 변환
df['재배기간_1'] = df['재배기간_1'].astype(int)
df['재배기간_2'] = df['재배기간_2'].astype(int)

# 수확기간
# 월/일을 datetime 객체로 변환
harvest_start_date_1 = convert_to_datetime(harvest_start_month_day_1)
harvest_end_date_1 = convert_to_datetime(harvest_end_month_day_1)

harvest_start_date_2 = convert_to_datetime(harvest_start_month_day_2)
harvest_end_date_2 = convert_to_datetime(harvest_end_month_day_2)

# 재배기간 컬럼 생성 (월일 기준으로 비교)
df['수확기간_1'] = (df['날짜'].dt.strftime('%m-%d') >= harvest_start_date_1.strftime('%m-%d')) & (df['날짜'].dt.strftime('%m-%d') <= harvest_end_date_1.strftime('%m-%d'))
df['수확기간_2'] = (df['날짜'].dt.strftime('%m-%d') >= harvest_start_date_2.strftime('%m-%d')) & (df['날짜'].dt.strftime('%m-%d') <= harvest_end_date_2.strftime('%m-%d'))

# 재배기간 컬럼을 True/False에서 1/0으로 변환
df['수확기간_1'] = df['수확기간_1'].astype(int)
df['수확기간_2'] = df['수확기간_2'].astype(int)


# 작물 생장과 관련된 컬럼 추가
# df['기온_18_20도'] = ((df['평균기온(℃)'] >= 18) & (df['평균기온(℃)'] <= 20)).astype(int)
# df['기온_15_18도'] = ((df['평균기온(℃)'] >= 15) & (df['평균기온(℃)'] <= 18)).astype(int)
df['기온_15_20도'] = ((df['평균기온(℃)'] >= 15) & (df['평균기온(℃)'] <= 20)).astype(int)

df['기온_영하3도이하'] = ((df['최저기온(℃)'] <= -3)).astype(int)
# df['기온_영하8도이하'] = ((df['최저기온(℃)'] <= -8)).astype(int)


# 새로운 그룹핑 기준 생성 (3월 기준 리셋)
df['누적그룹'] = np.where(df['월'] < 3, df['연도'] - 1, df['연도'])

# 재배기간 내 강수량
# 재배기간 내 일 강수량
df['재배기간_1_강수량'] = df.apply(lambda x: x['강수량(mm)'] if x['재배기간_1'] else 0, axis=1)
df['재배기간_2_강수량'] = df.apply(lambda x: x['강수량(mm)'] if x['재배기간_2'] else 0, axis=1)

# 재배기간 내 총 누적 강수량
df['재배기간_1_누적강수량'] = df.groupby('연도')['재배기간_1_강수량'].cumsum()
df['재배기간_2_누적강수량'] = df.groupby('누적그룹')['재배기간_2_강수량'].cumsum()

# 재배기간 누적 일수 계산
df['재배기간_1_누적일수'] = df.groupby('연도')['재배기간_1'].cumsum()
df['재배기간_2_누적일수'] = df.groupby('연도')['재배기간_2'].cumsum()

# 재배기간 내 당일까지 평균 강수량 0으로 나누는 문제 아직 있음
df['재배기간_1_평균강수량'] = np.where(
    df['재배기간_1_누적일수'] == 0, 0,  # 누적일수가 0일 경우 평균 강수량 0으로 설정
    df['재배기간_1_누적강수량'] / df['재배기간_1_누적일수']
)
df['재배기간_2_평균강수량'] = np.where(
    df['재배기간_2_누적일수'] == 0, 0,  # 누적일수가 0일 경우 평균 강수량 0으로 설정
    df['재배기간_2_누적강수량'] / df['재배기간_2_누적일수']
)


# 재배기간 내 생장 관련
df['재배기간_1_기온_15_20도'] = df.apply(lambda x: x['기온_15_20도'] if x['재배기간_1'] else 0, axis=1)
df['재배기간_2_기온_15_20도'] = df.apply(lambda x: x['기온_15_20도'] if x['재배기간_2'] else 0, axis=1)
df['재배기간_1_기온_영하3도이하'] = df.apply(lambda x: x['기온_영하3도이하'] if x['재배기간_1'] else 0, axis=1)
df['재배기간_2_기온_영하3도이하'] = df.apply(lambda x: x['기온_영하3도이하'] if x['재배기간_2'] else 0, axis=1)

# 재배기간 내 생장 관련 누적
df['재배기간_1_누적기온_15_20도'] = df.groupby('누적그룹')['재배기간_1_기온_15_20도'].cumsum()
df['재배기간_2_누적기온_15_20도'] = df.groupby('누적그룹')['재배기간_2_기온_15_20도'].cumsum()
df['재배기간_1_누적기온_기온_영하3도이하'] = df.groupby('누적그룹')['재배기간_1_기온_영하3도이하'].cumsum()
df['재배기간_2_누적기온_기온_영하3도이하'] = df.groupby('누적그룹')['재배기간_2_기온_영하3도이하'].cumsum()


df = df.fillna(0)
df
```
