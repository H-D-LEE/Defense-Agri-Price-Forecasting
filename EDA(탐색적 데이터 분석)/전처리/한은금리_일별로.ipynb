```python
df = pd.read_csv('C:\\Users\\acorn\\Desktop\\한국은행기준금리_변경일자.csv', encoding = 'ms949')
df
```

```python
df['날짜'] = df['년도'].astype(str) + '-' + df['날짜'].str.extract(r'(\d+)\s*월\s*(\d+)\s*일').apply(lambda x: f"{x[0]}-{x[1]}", axis=1)
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y-%m-%d')
df
```

```python
date_range = pd.date_range(start='2014-01-01', end='2024-10-31', freq='D')
daily_df = pd.DataFrame(date_range, columns=['날짜'])
daily_df
```

```python
daily_df.info()
```

```python
df_daily_geumri = pd.merge(daily_df, df, on='날짜', how='outer')
df_daily_geumri_filled = df_daily_geumri.ffill()
df_daily_geumri_filled
```

```python
df_daily_geumri_filled = df_daily_geumri_filled.rename(columns={'금리':'한은금리'})
df_haneungeumri = df_daily_geumri_filled[['날짜', '한은금리']]
df_haneungeumri
```
