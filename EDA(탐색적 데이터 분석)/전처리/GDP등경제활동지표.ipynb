```python
df_earn_from2014 = df_earn.iloc[19:]
df_earn_from2014
```

```python
df_earn_from2014_numeric = df_earn_from2014.applymap(pd.to_numeric, errors='coerce')
df_earn_from2014_numeric
```

```python
df_earn_from2014_numeric_10times = df_earn_from2014_numeric.applymap(lambda x: x * 10)
df_earn_from2014_numeric_10times
```

```python
df_earn_from2014_numeric_10times = df_earn_from2014_numeric_10times.fillna(0)
df_earn_from2014_numeric_10times
```

```python
df_earn_from2014_numeric_10times = df_earn_from2014_numeric_10times.astype('float').astype('int')
df_earn_from2014_numeric_10times
```

```python
df = df_earn_from2014_numeric_10times
df
```

```python
df = df.reset_index()
df
```

```python
df = df.rename(columns={'index':'기간'})
df
```

```python
# 날짜 범위 생성 (2014년 1월 1일부터 2024년 10월 31일까지)
dates = pd.date_range(start='2014-01-01', end='2024-10-31', freq='D')

# 빈 데이터프레임 생성 (일별 날짜 인덱스)
df_full = pd.DataFrame(index=dates)

# 분기별 데이터를 일별로 나누기 위한 함수
def fill_quarter_data(row):
    # 분기 시작일과 종료일 계산
    start_date = pd.to_datetime(row['기간'].split('/')[0] + '-' + row['기간'].split('/')[1].replace('Q', ''))
    end_date = start_date + pd.DateOffset(months=3) - pd.Timedelta(days=1)
    
    # 분기 내에 속하는 날짜들
    quarter_dates = pd.date_range(start=start_date, end=end_date, freq='D')
    
    # 일별로 값을 채워넣음
    for date in quarter_dates:
        df_full.loc[date, '농림어업'] = row['농림어업']
        df_full.loc[date, '국내총생산(시장가격, GDP)'] = row['국내총생산(시장가격, GDP)']
        df_full.loc[date, '국내총소득(GDI)'] = row['국내총소득(GDI)']
        df_full.loc[date, '국민총소득(GNI)'] = row['국민총소득(GNI)']

# 각 분기 데이터를 일별로 변환
df.apply(fill_quarter_data, axis=1)
```
