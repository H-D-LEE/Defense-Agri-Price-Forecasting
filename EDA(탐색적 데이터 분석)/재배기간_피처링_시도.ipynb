```python
# 재배기간 피처 엔지니어링
# 재배 시작일과 수확일 (월/일 기준으로 설정, 연도는 고려하지 않음)
# 감자
plant_start_month_day_1 = '03-01'
harvest_end_month_day_1 = '08-31'

plant_start_month_day_2 = '07-11'
harvest_end_month_day_2 = '12-21'

# 대파
plant_start_month_day_1 = '06-01'
harvest_end_month_day_1 = '11-01'

plant_start_month_day_2 = '07-11'
harvest_end_month_day_2 = '12-21'

# 연도를 임의로 지정 (예: 2000년)하여 월/일을 날짜로 변환
def convert_to_datetime(month_day_str):
    return pd.to_datetime(f"2000-{month_day_str}", format="%Y-%m-%d")

# 월/일을 datetime 객체로 변환
plant_start_date_1 = convert_to_datetime(plant_start_month_day_1)
harvest_end_date_1 = convert_to_datetime(harvest_end_month_day_1)

plant_start_date_2 = convert_to_datetime(plant_start_month_day_2)
harvest_end_date_2 = convert_to_datetime(harvest_end_month_day_2)

# 재배기간 컬럼 생성 (월일 기준으로 비교)
df['재배기간_1'] = (df['날짜'].dt.strftime('%m-%d') >= plant_start_date_1.strftime('%m-%d')) & (df['날짜'].dt.strftime('%m-%d') <= harvest_end_date_1.strftime('%m-%d'))
df['재배기간_2'] = (df['날짜'].dt.strftime('%m-%d') >= plant_start_date_2.strftime('%m-%d')) & (df['날짜'].dt.strftime('%m-%d') <= harvest_end_date_2.strftime('%m-%d'))

# 재배기간 컬럼을 True/False에서 1/0으로 변환
df['재배기간_1'] = df['재배기간_1'].astype(int)
df['재배기간_2'] = df['재배기간_2'].astype(int)

# 연도 컬럼 생성(연도별로 누적량 새로 시작해야 해서)
df['연도'] = df['날짜'].dt.year
```

```python
# 재배기간 일 강수량
# 밀양
df['일 강수량_밀양_재배기간_1'] = df.apply(lambda x: x['일 강수량_밀양'] if x['재배기간_1'] else 0, axis=1)
df['일 강수량_밀양_재배기간_2'] = df.apply(lambda x: x['일 강수량_밀양'] if x['재배기간_2'] else 0, axis=1)
# 대전
df['일 강수량_대전_재배기간_1'] = df.apply(lambda x: x['일 강수량_대전'] if x['재배기간_1'] else 0, axis=1)
df['일 강수량_대전_재배기간_2'] = df.apply(lambda x: x['일 강수량_대전'] if x['재배기간_2'] else 0, axis=1)
# 광주
df['일 강수량_광주_재배기간_1'] = df.apply(lambda x: x['일 강수량_광주'] if x['재배기간_1'] else 0, axis=1)
df['일 강수량_광주_재배기간_2'] = df.apply(lambda x: x['일 강수량_광주'] if x['재배기간_2'] else 0, axis=1)
# 인제
df['일 강수량_인제_재배기간_1'] = df.apply(lambda x: x['일 강수량_인제'] if x['재배기간_1'] else 0, axis=1)
df['일 강수량_인제_재배기간_2'] = df.apply(lambda x: x['일 강수량_인제'] if x['재배기간_2'] else 0, axis=1)
# 동두천
df['일 강수량_동두천_재배기간_1'] = df.apply(lambda x: x['일 강수량_동두천'] if x['재배기간_1'] else 0, axis=1)
df['일 강수량_동두천_재배기간_2'] = df.apply(lambda x: x['일 강수량_동두천'] if x['재배기간_2'] else 0, axis=1)

# 재배기간 누적 강수량 계산
# 밀양
df['누적_강수량_밀양_재배기간_1'] = df.groupby('연도')['일 강수량_밀양_재배기간_1'].cumsum()
df['누적_강수량_밀양_재배기간_2'] = df.groupby('연도')['일 강수량_밀양_재배기간_2'].cumsum()
# 대전
df['누적_강수량_대전_재배기간_1'] = df.groupby('연도')['일 강수량_대전_재배기간_1'].cumsum()
df['누적_강수량_대전_재배기간_2'] = df.groupby('연도')['일 강수량_대전_재배기간_2'].cumsum()
# 광주
df['누적_강수량_광주_재배기간_1'] = df.groupby('연도')['일 강수량_광주_재배기간_1'].cumsum()
df['누적_강수량_광주_재배기간_2'] = df.groupby('연도')['일 강수량_광주_재배기간_2'].cumsum()
# 인제
df['누적_강수량_인제_재배기간_1'] = df.groupby('연도')['일 강수량_인제_재배기간_1'].cumsum()
df['누적_강수량_인제_재배기간_2'] = df.groupby('연도')['일 강수량_인제_재배기간_2'].cumsum()
# 동두천
df['누적_강수량_동두천_재배기간_1'] = df.groupby('연도')['일 강수량_동두천_재배기간_1'].cumsum()
df['누적_강수량_동두천_재배기간_2'] = df.groupby('연도')['일 강수량_동두천_재배기간_2'].cumsum()

# 재배기간 누적 일수 계산
# 밀양
df['누적_일수_밀양_재배기간_1'] = df.groupby('연도')['재배기간_1'].cumsum()
df['누적_일수_밀양_재배기간_2'] = df.groupby('연도')['재배기간_2'].cumsum()
# 대전
df['누적_일수_대전_재배기간_1'] = df.groupby('연도')['재배기간_1'].cumsum()
df['누적_일수_대전_재배기간_2'] = df.groupby('연도')['재배기간_2'].cumsum()
# 광주
df['누적_일수_광주_재배기간_1'] = df.groupby('연도')['재배기간_1'].cumsum()
df['누적_일수_광주_재배기간_2'] = df.groupby('연도')['재배기간_2'].cumsum()
# 인제
df['누적_일수_인제_재배기간_1'] = df.groupby('연도')['재배기간_1'].cumsum()
df['누적_일수_인제_재배기간_2'] = df.groupby('연도')['재배기간_2'].cumsum()
# 동두천
df['누적_일수_동두천_재배기간_1'] = df.groupby('연도')['재배기간_1'].cumsum()
df['누적_일수_동두천_재배기간_2'] = df.groupby('연도')['재배기간_2'].cumsum()

# 재배기간 평균 강수량 구하기
# 밀양
df['평균_강수량_밀양_재배기간_1'] = df['누적_강수량_밀양_재배기간_1'] / df['누적_일수_밀양_재배기간_1']
df['평균_강수량_밀양_재배기간_2'] = df['누적_강수량_밀양_재배기간_2'] / df['누적_일수_밀양_재배기간_2']
# 대전
df['평균_강수량_대전_재배기간_1'] = df['누적_강수량_대전_재배기간_1'] / df['누적_일수_대전_재배기간_1']
df['평균_강수량_대전_재배기간_2'] = df['누적_강수량_대전_재배기간_2'] / df['누적_일수_대전_재배기간_2']
# 광주
df['평균_강수량_광주_재배기간_1'] = df['누적_강수량_광주_재배기간_1'] / df['누적_일수_광주_재배기간_1']
df['평균_강수량_광주_재배기간_2'] = df['누적_강수량_광주_재배기간_2'] / df['누적_일수_광주_재배기간_2']
# 인제
df['평균_강수량_인제_재배기간_1'] = df['누적_강수량_인제_재배기간_1'] / df['누적_일수_인제_재배기간_1']
df['평균_강수량_인제_재배기간_2'] = df['누적_강수량_인제_재배기간_2'] / df['누적_일수_인제_재배기간_2']
# 동두천
df['평균_강수량_동두천_재배기간_1'] = df['누적_강수량_동두천_재배기간_1'] / df['누적_일수_동두천_재배기간_1']
df['평균_강수량_동두천_재배기간_2'] = df['누적_강수량_동두천_재배기간_2'] / df['누적_일수_동두천_재배기간_2']

df.fillna(0, inplace=True) # 0으로 나누면 NaN 됨

df
```

# SDG 사용

```python
# StandardScaler 사용한 SGD(확률적 경사하강법)로 학습
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# from sklearn.preprocessing import StandardScaler # 표준화스케일링
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
print(X_train)

```

```python
# from sklearn import linear_model # SGD(확률적 경사하강법)로 학습시키기
model = linear_model.SGDRegressor(verbose=1)
model.fit(X_train, np.array(y_train).ravel())
```

```python
X.columns
```

```python
# 회귀식 기울기, 절편 확인
print('model.coef_ :', model.coef_) 
print('model.intercept_ :', model.intercept_)
print()
coef = pd.Series(data=np.round(model.coef_, 2), index=X.columns)
print(coef)
print()

# 평가 지표
# from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print('mse :', mse)
print('rmse :', rmse)
print('r2_score :', r2_score(y_test, y_pred))
```
