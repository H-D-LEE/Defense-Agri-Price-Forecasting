```python
import pandas as pd
```

```python
df_weather = pd.read_csv('C:\\Users\\acorn\\Desktop\\날씨데이터.csv', encoding = 'ms949')
df_weather
```

```python
df_5loc_weather = df_weather.loc[df_weather['지점번호'].isin([288, 156, 133, 98, 211])]
df_5loc_weather
```

```python
df_5loc_weather['지역'] = df_5loc_weather['지점번호'].map({288: '밀양', 156: '광주', 133: '대전', 98: '동두천', 211: '인제'})
df_5loc_weather
```

```python
df = df_5loc_weather
df
```

```python
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df
```

```python
df.info()
```

```python
df['날짜'] = pd.to_datetime(df['관측일'])
df
```

```python
df_loc_weather = df[['날짜', '지역', '일 평균기온', '최고기온', '최저기온', '일 평균 상대습도', '일 강수량', '최대풍속']]
df_loc_weather
```
