```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
df감자=pd.read_excel("C:/Users/kitcoop/Desktop/출하량_가락/고구마_출하량.xlsx")
df1=pd.read_excel("C:/Users/kitcoop/Desktop/데이터합.xlsx")
data=pd.merge(df감자,df1,left_on='DATE',right_on='분류',how='inner')
columns_to_select = ['일 평균기온_인제', '일 평균기온_동두천',
                     '일 평균기온_대전', '일 평균기온_밀양', '일 평균기온_광주'
                     ,'총반입량']
data1 = data[columns_to_select]
correlation = data1.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, center=0)
plt.title('Correlation Heatmap')
plt.show()
```
