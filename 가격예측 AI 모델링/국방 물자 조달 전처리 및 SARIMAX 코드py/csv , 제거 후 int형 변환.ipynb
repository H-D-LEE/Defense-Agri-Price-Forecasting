```python
for file_path in file_list:
    # 파일 이름 추출 (예: 감자_중품_100g_일별소매가)
    file_name = os.path.splitext(os.path.basename(file_path))[0]
    
    try:
        # 데이터 로드
        data = pd.read_csv(file_path, encoding='cp949')
        
        # 쉼표 포함된 데이터만 변환할 컬럼 지정
        columns_to_convert = ['평균', '전통시장', '유통업체']
        
        # 각 컬럼에 대해 쉼표가 포함된 값만 변환
        for col in columns_to_convert:
            if col in data.columns:  # 해당 컬럼이 존재할 경우만 처리
                data[col] = data[col].apply(
                    lambda x: int(x.replace(',', '')) if isinstance(x, str) and ',' in x else x
                )
        
        # 변환된 데이터를 저장 (Optional: 변경된 내용을 덮어쓰기 위해 같은 파일에 저장)
        output_path = f"C:/Users/acorn/Documents/python/일별소매가/processed_{file_name}.csv"
        data.to_csv(output_path, index=False, encoding='cp949')
        
        print(f"파일 처리 완료: {file_path} -> {output_path}")
    
    except Exception as e:
        print(f"파일 처리 중 오류 발생: {file_path}, 오류: {e}")
```
