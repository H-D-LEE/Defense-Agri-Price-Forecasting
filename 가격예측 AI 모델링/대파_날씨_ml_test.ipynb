```python
df_daepa = df_daepa[['날짜', '평균', '전통시장', '유통업체']]
df_daepa
```

```python
df_weather = pd.read_pickle("C:\\Users\\acorn\\Desktop\\weather_for_ml_test.pkl")
df_weather.columns = ['날짜', '지역', '평균기온', '최고기온', '최저기온', '평균상대습도', '강수량', '최대풍속']
df_weather
```

```python
df_weather_pivot = df_weather.pivot_table(index='날짜', columns='지역', values=['평균기온', '최고기온', '최저기온', '평균상대습도', '강수량', '최대풍속'])
df_weather_pivot
```

```python
df_weather_pivot.columns = [f"{col[1]}_{col[0]}" for col in df_weather_pivot.columns]
df_weather_pivot
```

```python
# df_weather_pivot.to_pickle("weather_pivot.pkl")
```

```python
df_daepa
```

```python
df_daepa_X = pd.merge(df_daepa, df_weather_pivot, on='날짜', how='left')
df_daepa_X
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic' # 한글폰트 지정
plt.rcParams['axes.unicode_minus']=False # 한글폰트 사용 시 -기호 깨지는 문제 해결
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(df_daepa_X.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```
